```python
import numpy as np
import numpy.random as nr
import random

total_episode = 50
learning_rate = 0.1
total_reward = 0
epsilon = 0.1
Q = np.ones([7,2])
goal_pos = 6
hole_pos = 0
actions = [0, 1]
movement = [-1, 1] # -1: move left, 1: move right
policy = [0.5, 0.5]
discount_factor = 0.9

POLICY_GREEDY = 1
POLICY_SAMPLING = 2
default_policy = POLICY_GREEDY #POLICY_SAMPLING
```

```python
def hidden_reward(pos):
    if pos == goal_pos:
        return 1
    if pos == hole_pos:
        return -1
    else:
        return 0
```

```python
for i in range(total_episode):
    curr_pos = 3
    print('----------------------------- Restart of the game -----------------------')
    for j in range(100):
        policy[0] = Q[curr_pos-1][0] / (Q[curr_pos-1][0] + Q[curr_pos+1][1])
        policy[1] = Q[curr_pos+1][1] / (Q[curr_pos-1][0] + Q[curr_pos+1][1])
        
        if default_policy == POLICY_GREEDY: #Epsilon greedy
            if nr.rand() < epsilon: #Exploration (Random movement)
                a = random.randrange(0, len(actions))
                print("Exploration choice in Epsilon greedy is ", a)
            else: #Exploitation
                a = np.argmax(policy)
                print("Exploitation choice in Epsilon greedy is ", a)
        else: # Epsilon sampling
            if nr.rand() < epsilon: #Exploration (Random movement)
                a = random.randrange(0, len(actions))
                print("Exploration choice in Epsilon sampling is ", a)
            else: #Sampling
                a = np.random.choice(actions, p=policy)
                print("Exploitation choice in Epsilon sampling is ", a)
        print("Current pos is", curr_pos, '. Selected action is ', movement[a])
        
        next_pos = curr_pos + movement[a]
        print("Now, our agent is at ", next_pos)
        
        reward = hidden_reward(next_pos)
        print("True reward of the action", a, 'is ', reward)
        
        np.maximum(Q[next_pos][0], Q[next_pos][1])
        
        Q[curr_pos][a] = Q[curr_pos][a] + learning_rate*(reward + discount_factor * np.maximum(Q[next_pos][0], Q[next_pos][1])-Q[curr_pos,a])
        
        print('Updated Q[curr_pos][a] is ',Q[curr_pos][a])
        print(Q)
        
        curr_pos = next_pos
        
        if reward == -1:
            break
        if reward == 1:
            total_reward += reward
            break
        print("Total success ratio is ", (total_reward/total_episode))
```
