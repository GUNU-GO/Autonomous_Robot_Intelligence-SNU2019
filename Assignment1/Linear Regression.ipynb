```python
import numpy as np
import matplotlib.pyplot as plt#시각화를 위해 import

data = np.loadtxt('data.csv', delimiter=',')

x_data = data[:,0]
y_data = data[:,1]

#Data 분포 확인용 plot
plt.scatter(x_data, y_data)
plt.show()

#Hyper parameter 설정
w = 0.1
b = 0.1
step = 0.0001
epoch = 200
n = float(len(x_data)) 

#Train
for i in range(epoch): 
    y = w*x_data + b  
    grad_w = (-1/n) * sum(x_data * (y_data - y - b)) #MSE를 미분하면 좌측과 같은 결과가 나옴
    grad_b = (-1/n) * sum(y_data - y - b)
    
    w = w - step * grad_w #Gradient descent method 적용
    b = b - step * grad_b  

#Result plot
y = w*x_data + b
plt.scatter(x_data, y_data)
x_list=np.arange(20,70,1)

for i in range(len(x_data)):
    plt.plot(x_data,y,color='red')
plt.show()
print("After training w : {} , b : {}".format(w,b))
```
