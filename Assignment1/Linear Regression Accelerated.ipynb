```python
import numpy as np
import matplotlib.pyplot as plt

data = np.loadtxt('data.csv', delimiter=',')

x_data = data[:,0]
y_data = data[:,1]

plt.scatter(x_data, y_data)
plt.show()

w = 0.1
b = 0.1
step = 0.0001
epoch = 200
m1 = 0.1
m2 = 0.1
n = float(len(x_data)) 

for i in range(epoch): 
    y = w*x_data + b  
    grad_w = (-1/n) * sum(x_data * (y_data - y - b))  
    grad_b = (-1/n) * sum(y_data - y - b)
    
    m1 = 0.9*m1 +step*grad_w #moment 도입, u=0.9로 설정
    m2 = 0.9*m2 +step*grad_b
    
    w = w - m1 #moment가 추가된 Gradient descent method
    b = b - m2
    
y = w*x_data + b
plt.scatter(x_data, y_data)
x_list=np.arange(20,70,1)

for i in range(len(x_data)):
    plt.plot(x_data,y,color='red')
plt.show()
print("After training w : {} , b : {}".format(w,b))
```
