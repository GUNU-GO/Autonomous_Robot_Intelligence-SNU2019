```python
import numpy as np
import matplotlib.pyplot as plt

def sigmoid(x):
    return 1/(1+np.exp(-x))

data = np.loadtxt('iris.csv', delimiter=',')
x1_data = data[:,0]
x2_data = data[:,1]
y_data = data[:,2]

for i in range(len(x1_data)):
    if y_data[i] == 1:
        plt.scatter(x1_data[i], x2_data[i], color="orange") 
    else:
        plt.scatter(x1_data[i], x2_data[i], color="steelblue") 
plt.show()

w1 = 0
w2 = 0
b = 2
step = 0.1
epoch = 100
m1=0.1
m2=0.1
m3=0.1
w1_list=[]
b_list=[]
x2_list=[]

n = float(len(x1_data)) 

for i in range(epoch): 
    y = w1*x1_data + w2*x2_data + b  
    grad_w1 =(1/n)*sum(x1_data*(sigmoid(y)-y_data)) 
    grad_w2 =(1/n)*sum(x2_data*(sigmoid(y)-y_data))
    grad_b =(1/n)*sum(sigmoid(y)-y_data)
    
    m1 = 0.9*m1+step*grad_w1 #moment 도입, u=0.9로 설정
    m2 = 0.9*m2+step*grad_w2
    m3 = 0.9*m3+step*grad_b
    
    w1 = w1 - m1 #moment가 추가된 Gradient descent method
    w2 = w2 - m2
    b = b - m3 
    
    w1_list.append(w1)
    b_list.append(b)
                         
y = w1*x1_data + w2*x2_data + b  
x = np.arange(4,8,0.1)
for i in range(len(x1_data)):
    if y_data[i] == 1:
        plt.scatter(x1_data[i], x2_data[i], color="orange") 
    else:
        plt.scatter(x1_data[i], x2_data[i], color="steelblue") 

for i in range(40):
    x2_list.append(-(w1*x[i]-b)/w2)
    
plt.plot(x,x2_list, color='red')
plt.show()
print("After training w1 : {} , w2 : {}, b : {}".format(w1,w2,b))
```
