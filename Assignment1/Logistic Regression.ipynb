```python
import numpy as np
import matplotlib.pyplot as plt #시각화를 위해 import

def sigmoid(x):
    return 1/(1+np.exp(-x))

#Data import
data = np.loadtxt('iris.csv', delimiter=',')
x1_data = data[:,0]
x2_data = data[:,1]
y_data = data[:,2]

#Data 분포 확인용 plot
for i in range(len(x1_data)):
    if y_data[i] == 1:
        plt.scatter(x1_data[i], x2_data[i], color="orange") 
    else:
        plt.scatter(x1_data[i], x2_data[i], color="steelblue") 
plt.show()

#Hyper parameter 설정
w1 = 0.1
w2 = 0.1
b = 0.5
step = 0.1
epoch = 100
x2_list=[]
n = float(len(x1_data)) 

#Train
for i in range(epoch): 
    y = w1*x1_data + w2*x2_data + b  
    grad_w1 =(1/n)*sum(x1_data*(sigmoid(y)-y_data)) #Active function을 sigmoid로 두고 CE의 미분을 구하면 좌측과 같음
    grad_w2 =(1/n)*sum(x2_data*(sigmoid(y)-y_data))
    grad_b =(1/n)*sum(sigmoid(y)-y_data)

    w1 = w1 - step * grad_w1 #Gradient descent method 적용
    w2 = w2 - step * grad_w2
    b = b - step * grad_b  

#Result plot
y = w1*x1_data + w2*x2_data + b  
x = np.arange(4,8,0.1)
for i in range(len(x1_data)):
    if y_data[i] == 1:
        plt.scatter(x1_data[i], x2_data[i], color="orange") 
    else:
        plt.scatter(x1_data[i], x2_data[i], color="steelblue") 
for i in range(40):
    x2_list.append(-(w1*x[i]-b)/w2)
    
plt.plot(x,x2_list, color='red')
plt.show()
print("After training w1 : {} , w2 : {}, b : {}".format(w1,w2,b))
```
